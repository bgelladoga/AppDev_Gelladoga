excercise 2

```python
try:
    hours = float(input("Enter Hours: "))
    rate = float(input("Enter Rate: "))

    if hours <= 40:
        pay = hours * rate
    else:
        regular_hours = 40
        overtime_hours = hours - 40
        pay = (regular_hours * rate) + (overtime_hours * rate * 1.5)

    print("Pay:", pay)
except ValueError:
    print("Please enter valid numeric input for hours and rate.")
```

```python
try:
    score = float(input("Enter a score between 0 and 100: "))
    if 97 <= score <= 100:
        grade = 1.0
    elif 93 <= score < 97:
        grade = 1.25
    elif 89 <= score < 93:
        grade = 1.5
    elif 85 <= score < 89:
        grade = 1.75
    elif 81 <= score < 85:
        grade = 2.0
    elif 77 <= score < 81:
        grade = 2.25
    elif 74 <= score < 77:
        grade = 2.5
    elif 0 <= score < 74:
        grade = 3.0
    else:
        print("Score out of range. Please enter a score between 0 and 100.")
        exit()

    print("Grade:", grade)
except ValueError:
    print("Please enter a valid numeric score.")
```
